```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
URL = "https://www.snapdeal.com/search?keyword=perfume%20men&sort=rlvncy"
```

```python
r = requests.get(URL)
htmlcontent=r.content
htmlcontent
```

```python
soup =  BeautifulSoup(htmlcontent,"html.parser")
print(soup)
soup.prettify
```

```python
per_name = soup.find_all(class_="product-title")
print(per_name)
```

```python
per_name_list = []
for i in range(0,len(per_name)):
    per_name_list.append(per_name[i].get_text())
print(per_name_list)    
    
    
```

```python
price = soup.find_all(class_="lfloat product-price")
print(price)
```

```python
price_list = []
price_int=[]
for i in range(0,len(price)):
    price_list.append(price[i].get_text())
    p1=price_list[i][5:]
    price_int.append(p1)
print(price_int)
```

```python
reviews = soup.find_all(class_="product-rating-count")
print(reviews)
```

```python
reviews_list = []
for i in range(0,len(reviews)):
    reviews_list.append(reviews[i].get_text())
print(reviews_list)    
```

```python
final_review_list = []
reviews_int=[]
for i in range(0,len(reviews_list)):
    final_review_list.append(reviews_list[i])
    p1=final_review_list[i][1:-1]
    reviews_int.append(p1)
print(reviews_int)
```

```python
import pandas as pd
df = pd.DataFrame({"Perfume_name":per_name_list,"Perfume_price":price_int,"Reviews":reviews_int})
print(df)
```

```python
import os
os.getcwd()
df.to_csv("snapdeal_perfume.csv",index=False)
df1 = pd.read_csv("snapdeal_perfume.csv")
df1
```
